```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
titanic = pd.read_csv("titanic.csv.bz2")
titanic.shape
```

```python
titanic = titanic[~titanic.age.isna()]
titanic.shape
```

```python
titanic = titanic.dropna(subset = ["fare"])
titanic.shape
```

```python
x = np.array([1, 2, 3])
np.mean((x - np.mean(x))**2)
```

```python
np.var(x)
```

```python
_ = plt.hist(titanic.age,
        	edgecolor = "k",
        	bins = 30)
_ = plt.xlabel("Age")
_ = plt.ylabel("Count")
```

```python
_ = plt.hist(titanic.fare,
        	edgecolor = "k",
        	bins = 100)
_ = plt.xlabel("Fare")
_ = plt.ylabel("Count")
```
