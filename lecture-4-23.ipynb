```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
R = 1000
T = np.random.lognormal(2.83, 0.06, size=R)
_ = plt.hist(T, bins=30, edgecolor = "w")
```

```python
np.percentile(T, [2.5, 97.5])
```

```python
titanic = pd.read_csv("titanic.csv.bz2")
titanic.shape
```

```python
fare = titanic.fare
fare = fare[~fare.isna() & (fare > 0)].values
fare.shape
```

```python
np.mean(fare), np.std(fare), np.median(fare)
```

```python
f = np.random.lognormal(2.95, 0.95, size = 1291)
np.mean(f), np.std(f), np.median(f)
```

```python
avgs = []
R = 1000
for i in range(R):
	f = np.random.lognormal(2.95, 0.95, size = 1291)
	avg = np.mean(f)
	avgs.append(avg)
np.mean(avgs)
```

```python
_ = plt.hist(avgs, bins=30, edgecolor = "w")
```

```python
np.percentile(avgs, [2.5, 97.5])
```

```python
mu = np.mean(avgs)
sd = np.std(avgs)
mu, sd
```

```python
mu - 1.96*sd, mu + 1.96*sd
```
