### 2 Theoretical approach 

1.


The Central Limit Theorem (CLT) states that the sampling distribution of the sample means approaches a normal distribution as the sample size gets larger, regardless of the shape of the population distribution. This is assuming that the samples are independent and identically distributed. The expected value (mean) of the sampling distribution of the sample mean will be equal to the mean of the population, and the standard deviation of the sampling distribution (the standard error) will be equal to the standard deviation of the population divided by the square root of the sample size.

2.

```python
female_mu = np.mean(female_avg_sim_fares)
print(female_mu)
female_standard_dev = np.std(female_avg_sim_fares)
print(female_standard_dev)

# Calculating the standard error for both genders
standard_error_female = female_std / np.sqrt(466)
print(standard_error_female)
```

```python
male_mu = np.mean(male_avg_sim_fares)
print(male_mu)
male_standard_dev = np.std(male_avg_sim_fares)
print(male_standard_dev)
standard_error_male = male_std / np.sqrt(10000)
print(standard_error_male)
```

3.

```python
# Z-score for 95% confidence
z_score = 1.96

# Calculating the 95% confidence interval for females
confidence_interval_female = (mean_female_fare - z_score * standard_error_female, mean_female_fare + z_score * standard_error_female)
print(confidence_interval_female)
```

```python
# Calculating the 95% confidence interval for males
confidence_interval_male = (mean_male_fare - z_score * standard_error_male, mean_male_fare + z_score * standard_error_male)
print(confidence_interval_male)
```
