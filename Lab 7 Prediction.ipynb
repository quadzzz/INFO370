```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import statsmodels.formula.api as smf
import seaborn as sns
```

```python
wbcd = pd.read_csv("wdbc.csv.bz2", sep = ",")
pd.set_option('display.max_columns', None)
wbcd
```

```python
y = wbcd.diagnosis.values
```

2.

```python
wbcd.iloc[:,[2,3,4]]
```

```python
X = wbcd.iloc[:,[2,3,4]].values
X
```

3.

```python
from sklearn.linear_model import LogisticRegression

m = LogisticRegression()
```

4.

```python
_ = m.fit(X, y)
m.coef_, m.intercept_
```

5.

```python
m.predict_proba(X)
```

```python
phat = m.predict_proba(X)[: , 1]
phat
```

```python
plt.hist(phat)
plt.xlabel("Probability of Having Breast Cancer")
plt.ylabel("Occurances")
_ = plt.show()
```
