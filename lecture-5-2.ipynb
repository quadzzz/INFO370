### 2024-05-02 multiple regression

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
ih = pd.read_csv("income-height.csv.bz2", sep = "\t")
ih = ih.dropna(subset = ["income","afqt", "height", "education"])
ih.shape
```

```python
ih.head(5)
```

```python
import statsmodels.formula.api as smf
```

```python
m = smf.ols("income ~ education + afqt", data=ih).fit()
m.summary()
```

```python
m = smf.ols("income ~ education + afqt + height", data=ih).fit()
m.summary()
```

```python
yhat = m.predict()
np.sqrt(np.mean((ih.income - yhat)**2))
```
