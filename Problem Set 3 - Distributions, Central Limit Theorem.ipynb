3. The least unequal distribution is the linear scale, since it has the most symmetrical and narrow shape, indicating a smaller variance and less spread. The most unequal distribution would be the log-linear scale, since it has a long tail and a wider spread, indicating higher variance and potentially more skewness.

4. 

```python
sample_stats = []
for sample in samples_ln:
    mean = np.mean(sample)
    variance = np.var(sample)
    std_dev = np.std(sample)
    coefficient_of_variation = std_dev / mean
    sample_stats.append((mean, variance, coefficient_of_variation))
```

```python
for stats, params in zip(sample_stats, parameters):
    print(f"Log-normal distribution with μ={params[0]} and σ={params[1]}:")
    print(f"Sample Mean: {stats[0]}, Sample Variance: {stats[1]}, Coefficient of Variation: {stats[2]}\n")
```

1. Random Variable (RV) is a central concept that connects probability theory to statistics. In particular, it makes numbers out of events so that one can use the mathematical apparatus to analyze the random processes. In essence, it is a rule that assigns a number to each event in the sample space S. RV X is X : S → R where S is the sample space of the phenomenon we are analyzing (pg 39). To clarify, RV is a rule, it is not random.

2. μ = Σx * P(x)

```python
probability = p = 0.5
possible_values_X = [-1, 1]
ExpectedValue = EX = 0
for x in [-1, 1]:
    EX += p * x
print(EX)
```

VarX = E(X^2) - (EX)^2

```python
EX2 = 0
for x in [-1, 1]:
    EX2 += p * (x**2)

VarX = EX2 - EX**2
print(VarX)
```

Expectation is a property of the random variable, a precisely defined stochastic process. Sample mean is a property of sample. Even if expectation is sometimes called “mean”, it is important to realize that sample and RV have different properties. For instance, sample mean is random and it fluctuates depending on what is sampled. But expectation is constant and does not change (pg 44). The difference between the expected value and the sample mean is that the expected value is a theoretical concept based on the probability distribution of X, while the sample mean is a statistical measure calculated from observed data.

```python
# Compute and report the mean and variance of the R pair means
sample_mean_of_means = np.mean(pair_means)
print(sample_mean_of_means)
sample_variance_of_means = np.var(pair_means, ddof=1)
print(sample_variance_of_means)
```

```python
# Compute the expected value and variance of the pair means (theoretical)
expected_value = np.sum(np.multiply(values, probabilities))
print(expected_value)

variance_of_single = np.sum(np.multiply(np.square(values), probabilities)) - np.square(expected_value)
theoretical_variance_of_means = variance_of_single / S
print(theoretical_variance_of_means)
```

12. As the tuple size S increases, the shape of the histogram for the means of these tuples becomes more bell-shaped and symmetrical, which is indicative of the normal distribution. This occurs due to the Central Limit Theorem (CLT), which states that as the size of the sample (or in this case, the size of the tuple) increases, the distribution of the sample means tends toward a normal distribution, regardless of the shape of the population distribution, provided the number of samples/tuples is sufficiently large.

13. The histograms resemble a normal distribution as S grows due to the CLT. When moving from single values (S=1) to pairs (𝑆=2), the histogram starts to take on the bell shape because the average of two random variables is more likely to be near the mean than at the extremes. With S=1, the histogram only shows the two possible values, resulting in two peaks. With 𝑆=2
S=2, the means can be -1, 0, or 1, which starts to fill in the gaps and form a "U"-shaped histogram. As S increases, the possible average values fill in more of the gaps, and the distribution smooths out into a bell curve. The extremes become less likely because they would require all or most of the independent random variables to be at one extreme, which is a rare event.

14. Difference Between 𝑅 and S: 
- R is the number of repetitions or the number of tuples generated. It represents how many samples we are drawing from the distribution.
- S is the sample size or the number of random variables in each tuple. It represents how many values are being averaged together to calculate each mean.

Changing R affects the smoothness and granularity of the histogram. A larger R provides a more accurate and smoother histogram, as there are more data points to plot, which gives a clearer picture of the distribution of sample means. Changing S affects the shape of the histogram. As S increases, the histogram of the sample means tends toward a normal distribution due to the CLT, even if the underlying distribution of individual observations is not normal. The variance of the sample means also decreases with larger S, resulting in a narrower histogram.
